```python
import numpy as np
from numpy import linspace, array
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# значения параметров (классические)
sigma = 10.0
rho = 28.0 
beta = 8.0/3.0

# aттрактор Лоренца
def attr_lorenz(t, xyz):
    x, y, z = xyz
    x_dot = sigma*(y - x)
    y_dot = rho*x - y - x*z
    z_dot = x*y - beta*z
    return [x_dot, y_dot, z_dot]

#начальное условие
y0 = [0, 1, 0]

# с какого времени стартуем и до какого идём
t_span = [0, 40]

# t_eval - отрезок делим на 5000 частей
solution = solve_ivp(attr_lorenz, t_span, y0, t_eval = linspace(0, 40, 5000))

# Plot
fig = plt.figure()
AL = fig.gca(projection = '3d')

AL.plot(solution.y[0], solution.y[1], solution.y[2], 'orange')
AL.set_xlabel("X")
AL.set_ylabel("Y")
AL.set_zlabel("Z")
AL.set_title("Аттрактор Лоренца")

plt.show()
```
